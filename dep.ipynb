```python
import dtlpy as dl
```

```python
use_rc_env = False
# dl.logout()
if use_rc_env:
    dl.setenv('rc')
else:
    dl.setenv('prod')
if dl.token_expired():
    dl.login()
```

```python
if use_rc_env:
    project = dl.projects.get(project_id='c28721df-bd4d-44ae-90f2-5e5a41d6dec4')
else:
    project = dl.projects.get(project_name='HusamTesting')
```

```python
# Publish your DPK (Dataloop Package)
dpk = project.dpks.publish(
)

# Install or update the application
try:
    app = project.apps.get(app_name=dpk.display_name)
    app.dpk_version = dpk.version
    app.update()
except dl.exceptions.NotFound:
    print("installing ...")
    app = project.apps.install(dpk=dpk)
```

```python
video_item = dl.items.get(item_id='69a01279b212b051f7379017')
video_stream_url = f"https://gate.dataloop.ai/api/v1/items/{video_item.id}/stream"

model = dl.models.get(model_id='69a012d5375acb7b9fb1ebd3')  # or by model_id
dataset = video_item.dataset
# Create a prompt item in the model's dataset
prompt_item = dl.PromptItem(name='video-describe-test')
prompt_item.add(
    message={
        "role": "user",
        "content": [
            {
                "mimetype": dl.PromptType.TEXT,
                "value": f"Describe what is happening in this video [video_url]({video_stream_url})"
            }
        ]
    },
    model_info={
        "name": model.name,
        "model_id": model.id,
    }
)

# Upload the prompt item to the model's dataset
item = dataset.items.upload(prompt_item)
```
